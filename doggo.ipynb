```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('seaborn-dark')
```

```python
img = plt.imread('doggo.png')
img = img[:,:700,1]
plt.imshow(-1*img)
```

```python
n = 25
m = 35
pix = np.zeros([n,m])
for i in range(n):
    for j in range(m):
        pix[i][j] = int(np.round(np.mean(img[20*i:20*i+19,20*j:20*j+19]),1)*10)
plt.imshow(-1*pix)
```

```python
cdict = {'red':   ((0.0, 0.0, 0.),
                   (.33,0.34,0.34),
                   (.66,0.86,0.86),
                   (1.0, 1.0, 0.91)),

         'green': ((0.0, 0.0, 0.0),
                   (0.33, 0.22, 0.22),
                   (0.66, 0.57, 0.57),
                   (1.0, 1,.57)),

         'blue':  ((0.0, 0.0, 0.0),
                   (0.33, 0.05, 0.05),
                   (0.66, 0.12, 0.12),
                   (1.0, 1, .37))
        }

brown = mpl.colors.LinearSegmentedColormap('brown', cdict)
plt.register_cmap('brown',brown)
```

```python
plt.imshow(pix, cmap=brown)
```
